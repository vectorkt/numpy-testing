**Inteleg ce e un dot product, chiar pot sa il fac de mana. Spe exemplu, inteleg inmultirea a doua matrici de shape (1,2) dot (2,1) rezultatul ar fi (1,1)**

```python
import numpy as np

#matrix multiplication
matr1=np.random.random([1,2])
matr2=np.random.random([2,1])

print(matr1.shape)
#print(matr1)
print(matr2.shape)
#print(matr2)

matr3=np.dot(matr1,matr2)

print(matr3.shape)
```

**Pentru ca dot-ul clasic nu prea are success, am incercat cu tesor dot, care are succese variable. In urmatoarea, de exemplu, rezultatul pare sa functioneze**

```python
image_section=np.random.random([4,4,1])
filtr=np.random.random([4,4,1])

result=np.tensordot(image_section,filtr,axes=([1,0],[0,1]))

print(image_section.shape)
print(filtr.shape)
print(result.shape)
print(result)
```
